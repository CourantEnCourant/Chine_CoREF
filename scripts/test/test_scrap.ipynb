```python
import time

'''import docstring'''
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.common.by import By
```

```python
wd = webdriver.Chrome()
wd.implicitly_wait(5)
```

```python
site_start = 'https://chinadigitaltimes.net/chinese/704083.html'
wd.get(site_start)

urls = wd.find_elements(By.XPATH, '//*[@href]')
```

```python
attributes = [href.get_attribute('href') for href in urls]
for attribute in attributes:
    print(attribute)
```

```python
xpath_pattern = '//*[@href and not(contains(@href, "css") or contains(@href, "png") or contains(@href, "php"))]'
urls_filtered = wd.find_elements(By.XPATH, xpath_pattern)
```

```python
attributes = [href.get_attribute('href') for href in urls_filtered]
print(len(attributes))
for attribute in attributes:
    print(attribute)
```

```python
import time

for attribute in attributes:
    wd.get(attribute)
    print(f'Navigating to {attribute}')
    time.sleep(5)
```
